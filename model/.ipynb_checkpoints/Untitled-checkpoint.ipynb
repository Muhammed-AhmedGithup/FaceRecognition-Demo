```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import warnings
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
cropped_path='./dataset/cropped/'
```

```python
images_dir=[i.path for i in os.scandir(cropped_path)]
images_dir
```

```python
cerebity_name_class_dic={}
for imge_d in images_dir:
    name_cerebity=imge_d.split('/')[-1]
    files=[]
    for i in os.scandir(imge_d):
        files.append(i.path)
    cerebity_name_class_dic[name_cerebity]=files
```

```python
cerebity_name_class_dic
```

```python
classes_dict={}
count=0
for image_dir in images_dir:
    celebity_name=image_dir.split('/')[-1]
    classes_dict[celebity_name]=count
    count=count+1
```

```python
classes_dict
```

```python
import pywt
import cv2    

def w2d(img, mode='haar', level=1):
    imArray = img
    #Datatype conversions
    #convert to grayscale
    imArray = cv2.cvtColor( imArray,cv2.COLOR_RGB2GRAY )
    #convert to float
    imArray =  np.float32(imArray)   
    imArray /= 255;
    # compute coefficients 
    coeffs=pywt.wavedec2(imArray, mode, level=level)

    #Process Coefficients
    coeffs_H=list(coeffs)  
    coeffs_H[0] *= 0;  

    # reconstruction
    imArray_H=pywt.waverec2(coeffs_H, mode);
    imArray_H *= 255;
    imArray_H =  np.uint8(imArray_H)

    return imArray_H
```

```python
x=[]
y=[]
for class_name,files_train in cerebity_name_class_dic.items():
    for i in files_train:
        image_train=cv2.imread(i)
        image_train=cv2.resize(image_train,(50,50))
        imag_wv=w2d(image_train,'db1',5)
        imag_wv=cv2.resize(imag_wv,(50,50))
        combine=np.vstack((image_train.reshape(50*50*3,1),imag_wv.reshape(50*50,1)))
        x.append(combine)
        y.append(classes_dict[class_name])
        
```

```python
x=np.array(x).reshape(len(x),10000)
```

```python
x[0]
```

```python
x.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV ,train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model_params = {
    'svm': {
        'model': SVC(gamma='auto',probability=True),
        'params' : {
            'svc__C': [1,10,100,1000],
            'svc__kernel': ['rbf','linear']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'randomforestclassifier__n_estimators': [1,5,10]
        }
    },
    'logistic_regression' : {
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params': {
            'logisticregression__C': [1,5,10]
        }
    }
}
```

```python
estimators={}
scores=[]
for k,v in model_params.items():
    pip=make_pipeline(StandardScaler(),v['model'])
    estimete=GridSearchCV(estimator=pip,param_grid=v['params'],cv=5)
    estimete.fit(x_train,y_train)
    estimators[k]=estimete
    scores.append({
        'model':v['model'],
        'best_score':estimete.best_score_,
        'best_params':estimete.best_params_,
    })

df_scores=pd.DataFrame(scores,columns=['model','best_score','best_params'])
df_scores   
```

```python
def classify_image(image):
    image=cropped_2eys(image=image)
    image_face_color=cv2.resize(image,(50,50))
    image_wv=w2d(img=image,mode='db1',level=5)
    image_wv=cv2.resize(image_wv,(50,50))
    final=np.vstack((image_face_color.reshape(50*50*3,1),image_wv.reshape(50*50,1)))
    final=np.array(final).reshape(10000)
    return pipe.predict([final])[0]
```

```python
def cropped_2eys(image):
    face_cascade=cv2.CascadeClassifier('./opencv/haarcascades/haarcascade_frontalface_default.xml')
    eys_cascade=cv2.CascadeClassifier('./opencv/haarcascades/haarcascade_eye.xml')
    image=cv2.imread(image)
    gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray_image,1.3,5)
    for(x,y,w,h) in faces:
        gray_p=gray_image[y:y+h,x:x+w]
        face_color=image[y:y+h,x:x+w]
        eyes=eys_cascade.detectMultiScale(gray_p)
        if len(eyes)>=2:
            return face_color
        
```
